## Relatório de análise dos dados
<br>
<p> Após a leitura dos arquivos correspondentes, começamos o tratamento dos dados. Iniciamos com a limpeza dos valores ausentes da coluna “Year” do dataframe WorldCupMatches. </p>

```python
#Limpeza da coluna Year da tabela matches "a coluna possui valores ausentes"

matches.dropna(subset=['Year'], inplace=True)
```

<p> Após isso, criamos a coluna “TotalGoals” com todos os gols feitos em na partida definida pela linha do dataframe. </p>

```python
#Criando coluna com total de gol de uma determinada partida

matches['TotalGols'] = matches['Home Team Goals'] + matches['Away Team Goals']

matches.head()
```

<p> Nessa parte, verificamos quantos jogos cada time jogou sendo o “Home Team Name”, traduzido como time da casa. Entretanto, o termo foi usado de forma errônea, pois na procura pela quantidade de partidas de cada time é necessário possuir uma coluna especificando o país onde foi jogado. </p>
<p> Por isso quando realizamos a busca, muitos países que não sediaram a Copa do Mundo possuem partidas contabilizadas e, até mesmo, países que sediaram a Copa com números exuberantes de partidas. </p>

```python
#Verificando o nome do time da casa e a quantidade de jogos que teve no pais

matches['Home Team Name'].value_counts()
```

<p> Para a resolução deste problema, sugerimos a alteração dos nomes das colunas “Home Team Name” e “Away Team Name” para “Team Winner” e “Team Loser”, traduzido para time ganhador e time perdedor. Dessa forma, não haveria chance de interpretações erradas, já que, claramente, o time sediador da Copa do Mundo não jogou em todas as partidas realizadas. E, para ser possível a contabilização do número de partidas jogadas pelo país sediador, seria interessante a criação da coluna “Host Country”, traduzido para país sede, para comparação das colunas “Team Winner” e “Team Loser” para a contabilização da quantidade de vezes que cada um aparece. </p>
<p> Após isso, removemos todos os caracteres “rn”, “ “ “ (aspas duplas) e “>” da coluna “Home Team Name”, criamos um objeto de lista para separar os nomes errados dos nomes certos. </p>
<p> Também foi usada o método split para separarmos o que vinha depois do caractere “>”, pois o nome correto do time se encontrava ali. </p>

```python
#Remoção de todas das caracterias "rn,>" na coluna "Home Team Name"
names = matches[matches['Home Team Name'].str.contains('rn">')]['Home Team Name'].value_counts()

#Acriação de um objeto de lista e com nome de wrong
wrong = list(names.index)

#Realizando um split exatamente no caractere ">" para assim separar em duas posições
#e pegamos a segunda posição (que é o [1]) que é exatamente o nome do país sem os caracteres
correct = [name.split('>')[1] for name in wrong]
correct
```

<p> Fazemos algo parecido para os estádios que possuem caracteres especiais, como “Maracanã”, mudando o nome deles para algo mais simples e sem os caracteres. Declaramos uma variável para inserir os nomes errados, “old_name”, e uma nova para inserir os nomes certos, “new_name”. Após fazer isso, comparamos os nomes errados com os nomes corretos. </p>

```python
#Existencia de caracterias especiais como 'ã' está presente em toda tabela então criarei duas listas para separalas

old_name = ['Germany FR', 'Maracan� - Est�dio Jornalista M�rio Filho', 'Estadio do Maracana']
new_name = ['Germany', 'Maracan Stadium', 'Maracan Stadium']

wrong = wrong + old_name
correct = correct + new_name

#Aqui vemos como os dois ficam:
wrong, correct
```

<p> Após o tratamento descrito acima, finalmente colocamos os nomes corretos contidos na variável “correct” e no lugar dos nomes errados contidos na varíavel “wrong”, fazemos isso com todos os dataframes. </p>

```python
#Modificando a posição de todos os campos com caractere especial
#para os dados "corretos" na mesma posição
for index, wr in enumerate(wrong):
    world_cup = world_cup.replace(wrong[index], correct[index])

for index, wr in enumerate(wrong):
    matches = matches.replace(wrong[index], correct[index])

for index, wr in enumerate(wrong):
    players = players.replace(wrong[index], correct[index])

    display(matches)
```

<p> E conferimos a legibilidade dos dados para conferir se os caracteres “rn”, “ “ “ (aspas duplas) e “>” ainda existem nos dataframes. </p>

```python
#Testando a legibilidade dos dados após alterações
names = matches[matches['Home Team Name'].str.contains('rn">')]['Home Team Name'].value_counts()
names
```

<p> Realizamos um tratamento de NaN’s, “NotaNumber” traduzido para Não é um Número, e colocamos o caracter “-“ no lugar para que as células tenham valor. </p>

```python
#Realizando o tratamento de NAN's

players = players.replace([float('NaN')],'-')
matches = matches.replace([float('NaN')],'-')
world_cup = world_cup.replace([float('NaN')],'-')
```

<p> Como foi pedido, distribuímos os itens para análise de dados por dados informados nos dataframes. </p>

```python
# Seleções participantes

plt.figure(figsize=(15,4))
sns.barplot(x='Year',y='QualifiedTeams', data=world_cup)
plt.title("Quantidade de seleções participantes por copa do mundo",color='black')
plt.show()
```

```python
# Quantidade de partidas disputadas por copa do mundo

plt.figure(figsize=(15,4))
sns.pointplot(x='Year', y ='MatchesPlayed', data=world_cup, color="orange")
plt.grid(True,color="grey",alpha=.3)
plt.title("Quantidade de partidas por copa do mundo",color='black')
plt.show()
```

```python
# Média de gols por copa do mundo
# Crie um DataFrame com as colunas 'Year' e 'TotalGols'
data = pd.DataFrame({'Year': matches['Year'], 'TotalGols': matches['TotalGols']})

# Crie o gráfico de boxplot
plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='Year', y='TotalGols')
plt.title("Média de gols por Copa do Mundo", color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
# Quantidade de gols marcados por copa do mundo
plt.figure(figsize=(15,5))
sns.pointplot(x='Year', y ='GoalsScored', data=world_cup, color="green")
plt.grid(True,color="grey",alpha=.3)
plt.title("Quantidade de gols marcados por copa do mundo",color='black')
plt.show()
```

<p> Após as médias feitas, também realizamos apresentação de gráficos indicando dados interessantes contidos nos dataframes. </p>

```python
# Total de gols por seleção
gols_casa = matches.groupby("Home Team Name")["Home Team Goals"].sum().reset_index()
gols_casa.columns = ["selecao","gols"]
gols_fora = matches.groupby("Away Team Name")["Away Team Goals"].sum().reset_index()
gols_fora.columns = ["selecao","gols"]
gols = pd.concat([gols_casa, gols_fora],axis=0)
gols = gols.groupby("selecao")["gols"].sum().reset_index()
gols = gols.sort_values(by="gols",ascending =False)
gols["gols"] = gols["gols"].astype(int)

plt.figure(figsize=(15,5))
sns.barplot(x="selecao",y="gols", data=gols[:10])
plt.title("Seleções com maior número de gols marcados",color='black')
plt.show()
```

```python
# Aqui podemos visualizar com um gráfico os maiores campeões da história da Copa do Mundo
winner = world_cup['Winner'].value_counts()

plt.figure(figsize=(15,5))
sns.countplot(x='Winner', data=world_cup, order=winner.index)
plt.title("Maiores campeões",color='black')
plt.show()
```

```python
# Já aqui podemos visualizar com um gráfico os maiores vice-campeões da história da Copa do Mundo
runnerup = world_cup['Runners-Up'].value_counts()

plt.figure(figsize=(15,5))
sns.countplot(x='Runners-Up', data=world_cup, order=runnerup.index)
plt.title("Maiores vice-campeões",color='black')
plt.show()
```

```python
# Já aqui podemos visualizar com um gráfico aqueles que ficaram mais vezes na posição de terceiro lugares
# na história da Copa do Mundo
third = world_cup['Third'].value_counts()

plt.figure(figsize=(15,5))
sns.countplot(x='Third', data=world_cup, order=third.index)
plt.title("Maiores terceiro lugar",color='black')
plt.show()
```

<p> E, por fim, concatenamos as três colunas que estruturamos acima nos gráficos, formando uma tabela, transformamos os NaN’s presentes em 0 e declaramos todos os números como inteiros. </p>

```python
# Aqui, vamos concatenar as três colunas que estrutramos acima nos gráficos.
# Além disso, transformaremos todos os campos NAN's em 0 e transformaremos todos os valores em inteiros
teams = pd.concat([winner, runnerup, third], axis=1)
teams.fillna(0, inplace=True)
teams = teams.astype(int)
teams
```
